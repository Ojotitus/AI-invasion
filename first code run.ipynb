```python
#Label Encoding
cat_features = ["Pregnancies","Glucose","BloodPressure","SkinThickness","Insulin","BMI","DiabetesPedigreeFunction","Age","Outcome"]
for cat_feature in cat_features:
  df[f"{cat_feature}_cat"] = df[cat_feature].astype ("category")
  df[f"{cat_feature}_cat"] = df[f"{cat_feature}_cat"].cat.codes
```

```python
df.head(10)
```

```python
#Drop redundant features
df.drop(["Pregnancies","Glucose","BloodPressure","SkinThickness","Insulin","BMI","DiabetesPedigreeFunction","Age","Outcome"], axis = 1, inplace=True)
df.head(10)
```

```python
#Data segementation
y = df["Outcome_cat"] #Target
x = df.drop("Outcome_cat", axis = 1)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.3)
```

```python
#Loading the dataset into Linear Regression algorithm
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train, y_train)
LinearRegression()
```

```python
#Making predictions
reg.predict(x_test)
```

```python
#Model Evaluation
from sklearn.metrics import mean_absolute_error
y_predict = reg.predict (x_test)
print ("MAE", mean_absolute_error (y_test, y_predict))
```
